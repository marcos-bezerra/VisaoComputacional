```python
# %load imageblur.py
# Adding noise to the image    

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('../images/messi5_gray_saltpeper.jpg',0)
im = np.zeros(img.shape, np.uint8) # do not use original image it overwrites the image
# "Original"
plt.imshow(img)
```

```python
im = cv2.GaussianBlur(img,(3,3),0)
# "Gaussian 3x3*"
plt.imshow(im)
```

```python
im = cv2.GaussianBlur(img,(5,5),0)
#"Gaussian 5x5"
plt.imshow(im)
```

```python
im = cv2.medianBlur(img, 3)
#"Median 3x3"
plt.imshow(im)
```

```python
im = cv2.medianBlur(img, 5)
#"Median 5x5"
plt.imshow(im)
```

```python
im = cv2.blur(img,(3,3) )
#"Blur 3x3"
plt.imshow(im)
```

```python
im = cv2.blur(img, (5,5) )
# "Blur 5x5"
plt.imshow(im)
```
